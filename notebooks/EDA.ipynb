```python
import plotly.express as px

fig = px.histogram(df, x='model_year', nbins=25,
                   title='Distribution of Vehicle Model Years')
fig.show()
```

```python
fig = px.scatter(df, x='days_listed', y='price',
                 color='fuel', 
                 title='Price vs. Mileage by Fuel Type',
                 opacity=0.6)
fig.update_traces(marker=dict(size=5))
fig.show()
```

```python
fig = px.scatter_3d(df, x='type', y='condition',z='price',
                    color='fuel', size='price',

                    title='Vehicle price by type, fuel type, and condition')
fig.show()
```
